```python
import requests
import pandas as pd

url = "https://api.binance.com/api/v3/klines"
symbol = "BTCUSDT"
interval = "1d"
limit = 1000
end_time = None
all_data = []

while True:
    params = {
        "symbol": symbol,
        "interval": interval,
        "limit": limit,
        "endTime": end_time
    }
    response = requests.get(url, params=params)
    data = response.json()
    
    if not data:
        break
    
    all_data.extend(data)
    end_time = data[0][0]
    
    if len(all_data) >= 1400:
        break

df = pd.DataFrame(all_data, columns=['Open time', 'Open', 'High', 'Low', 'Close', 'Volume',
                                     'Close time', 'Quote asset volume', 'Number of trades',
                                     'Taker buy base volume', 'Taker buy quote volume', 'Ignore'])

df['Open time'] = pd.to_datetime(df['Open time'], unit='ms')
df['Close time'] = pd.to_datetime(df['Close time'], unit='ms')

df = df.iloc[::-1].reset_index(drop=True)

print(df)
```

```python
df = pd.DataFrame(data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume',
           'Close time', 'Quote asset volume', 'Number of trades',
           'Taker buy base volume', 'Taker buy quote volume', 'Ignore'])
df['Date'] = pd.to_datetime(df['Date'], unit='ms')
df
```

```python
df_filtered = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']].copy()
cols = ['Open', 'High', 'Low', 'Close', 'Volume']
df_filtered[cols] = df_filtered[cols].astype(float).round(2)
df_filtered['Date'] = df_filtered['Date'].dt.strftime('%Y-%m-%d')

df_filtered
```

```python
donchian_up = df_filtered['High'].rolling(20,min_periods=1).max()
donchian_low = df_filtered['Low'].rolling(20,min_periods=1).min()
donchian_mid = (donchian_up+donchian_low)/2
```

```python
cols_donchian = ['up','mid','low']
df_donchian = pd.DataFrame(columns=cols_donchian)
for col in df_donchian.columns:
    df_donchian['up']=donchian_up
    df_donchian['low']=donchian_low
    df_donchian['mid']=donchian_mid

df_donchian.dropna(inplace=True)
df_donchian.set_index(df_filtered['Date'].iloc[-len(df_donchian):], inplace=True)
df_donchian
```

```python
df_filtered['TrailingStop'] = None
for i in range(len(df_filtered)):
    if i ==0:
        df_filtered.loc[i, 'TrailingStop'] = donchian_mid.iloc[i]
    else:
            df_filtered.loc[i, 'TrailingStop'] = max(df_filtered.loc[i-1, 'TrailingStop'], donchian_mid.iloc[i])

df_filtered[['Date','Close','TrailingStop']]
```

```python
# Check if the closing price crosses the Donchian Channel
crossed = ((df_filtered['Close'] > donchian_up) | (df_filtered['Close'] < donchian_low)).any()

if crossed:
    print("The price crosses the Donchian Channel at least once.")
else:
    print("The price does not cross the Donchian Channel.")
```

```python
df_filtered['Returns'] = df_filtered['Close'].pct_change()
df_filtered.dropna()
vol_90d = df_filtered['Returns'].rolling(window=90).std()*np.sqrt(252)
df_filtered['Vol_90D'] = vol_90d
df_filtered
```
